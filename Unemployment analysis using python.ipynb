```python
data.columns= ["States", "Date", "Frequency",  
               "Estimate Unemployed Rate",  
               "Estimate Labour Participation Rate",  
                "longitude", "latitude"]  
```

```python
# ploting the heatmap
plot.style.use('seaborn-whitegrid')
plot.figure(figsize=(22,20))
sns.heatmap(data.corr())
plot.show()
```

```python
# ploting the histplot
data.columns=["States","Frequency","Estimate Unemployed Rate","Estimate Labour Participation Rate","longitude","latitude","Region"]
plot.title("Indian Unemployed")
sns.histplot(x="Estimate Unemployed Rate",hue="Region",data=data)
plot.show()
```
